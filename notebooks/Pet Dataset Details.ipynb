```python
import os
import matplotlib.pyplot as plt
import numpy as np
import cv2
from PIL import Image
```

```python
# Randomly select some images
IMAGE_DIR = "../data/oxford-iiit-pet/images"  # Change to the folder where images are stored
MASK_DIR = "../data/oxford-iiit-pet/annotations/trimaps"  # Change to the folder where annotations are stored

# Get a list of image and mask filenames
image_files = sorted([f for f in os.listdir(IMAGE_DIR) if f.endswith('.jpg')])
mask_files = sorted([f for f in os.listdir(MASK_DIR) if f.endswith('.png') and not f.startswith('.')])  # Segmentation masks

# Select a few samples to display
num_samples = 5
sample_indices = np.random.choice(len(image_files), num_samples, replace=False)
```

```python
mask = cv2.imread(os.path.join(MASK_DIR, mask_files[200]))
unique_pixel_values = np.unique(mask)
print("unique pixel values in mask: ", unique_pixel_values)
```

```python
# Plot images, their segmentation masks and contours
fig, axes = plt.subplots(num_samples, 3, figsize=(10, num_samples * 3))

for i, idx in enumerate(sample_indices):
    # Load image and mask
    image = cv2.imread(os.path.join(IMAGE_DIR, image_files[idx]))
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert to RGB for matplotlib

    mask = cv2.imread(os.path.join(MASK_DIR, mask_files[idx]), cv2.IMREAD_GRAYSCALE)

    # Threshold the mask to get a binary image
    _, binary_mask_2 = cv2.threshold(mask, 2, 255, cv2.THRESH_BINARY)

    # Find contours
    contours, _ = cv2.findContours(binary_mask_2, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)

    # Draw contours on the image
    contour_image = image.copy()
    cv2.drawContours(contour_image, contours, -1, (255, 0, 0), thickness=2)  # Blue color contours

    # Display image
    axes[i, 0].imshow(image)
    axes[i, 0].set_title(f"Image: {image_files[idx]}")
    axes[i, 0].axis("off")

    # Display mask
    axes[i, 1].imshow(mask, cmap="jet", alpha=0.3)  # Overlay with transparency
    axes[i, 1].set_title(f"Mask: {mask_files[idx]}")
    axes[i, 1].axis("off")

    # Display image with contours
    axes[i, 2].imshow(contour_image)
    # axes[i, 2].imshow(binary_mask)
    axes[i, 2].set_title(f"Contours: {mask_files[idx]}")
    axes[i, 2].axis("off")

plt.tight_layout()
plt.show()
```

# Get some data about the sizes

```python
unique_sizes = set()

for file_name in mask_files:
    with Image.open(os.path.join(MASK_DIR, file_name)) as img:
        unique_sizes.add(img.size)  # img.size returns (width, height)

# Convert set to sorted list (optional)
unique_sizes = sorted(unique_sizes)

print(unique_sizes)
```
